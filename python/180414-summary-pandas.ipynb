### Data Wrangling

```python
df1 = pd.DataFrame({'data1' : range(7), 'key' : list('bbacaab')})
df2 = pd.DataFrame({'data2' : range(20,23), 'key' : list('abd')})
print(df1)
print(df2)
```

```python
#can do inner,outer, left, and right joins
#can specify the columns with on=, left_on, etc
#can use index labels as keys
df1.merge(df2, on="key", how='inner')
```

```python
df1.append(df2)
```

```python
#good approach if all columns are the same
pd.concat([df1,df2])
```

### Normalizing data

```python
df1["strings"] = ["awkdkwaf", "JWJQFW", "  d", "a   ", "A", "akKlwA", "  l  "]
```

```python
#could also use an apply/applymap, note how we first need to call str
print(df1["strings"].str.lower())
print(df1["strings"].str.strip())
print(df1["strings"].str.startswith("a"))
print(df1["strings"].str.contains("a"))
```
